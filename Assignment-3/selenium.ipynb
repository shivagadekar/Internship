```python
import selenium
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException , NoSuchElementException
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome('C:/chrome_driver/chromedriver.exe')  # Loading Chrome Driver
driver.get('https://www.amazon.in/')  # opening web address in chrome driver browser
```

```python
product = driver.find_element(By.ID, "twotabsearchtextbox")  # Finding element with the help ID
product.send_keys('Mobiles')  # Passing "Laptop" as an search keyword
product.send_keys(Keys.RETURN)  # Pressing enter to load results
```

```python
page_results = driver.find_elements(By.XPATH, '//div[@class="s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis s-latency-cf-section s-card-border"]')
for i in page_results:
    # Name
    # print(i.text)
    print('----------------------------------')
    brand_name = i.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for _ in brand_name:
        print(_.text)
    # print((brand_name.text))
    # Splitting Name
    # name = brand_name.text.split(' ')[0]
    # print(name)
```

# Brand Names

```python
all_results = driver.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal"]')
for result in all_results:  # Looping for first 10 lapotps
    i = result.text
    b = i.split(' ')[0]
    print(b)
    print(i[len(b)+1:])
    print('---------------------------------------------------------------------------------------------------')
```

```python
# finding elements with price information and extracting prices
price_element = driver.find_elements(By.XPATH, '//span[@class="a-price-whole"]')
for _ in price_element:
    i = _.text 
    print(i)
```

```python
# Finding Expected Delivert
Exp_delivery_date = driver.find_elements(By.XPATH, '//span[@class="a-color-base a-text-bold"]')
for _ in Exp_delivery_date:
    i = _.text 
    print(i)
```
