```python
df['length'] = df.comment_text.str.len()
df.head(5)
```

```python
# Convert all comments to lower case
df['comment_text'] = df['comment_text'].str.lower()
```

```python
# Remove punctuation
df['comment_text'] = df['comment_text'].str.replace(r'[^\w\d\s]', ' ')
```

```python
# Replace whitespace between terms with a single space
df['comment_text'] = df['comment_text'].str.replace(r'\s+', ' ')
```

```python
#Remove leading and trailing whitespace
df['comment_text'] = df['comment_text'].str.replace(r'^\s+|\s+?$', '')
```

```python
# Replace email addresses with ' '
df['comment_text'] = df['comment_text'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$', ' ')
```

```python
# Replace URLs with ' '
df['comment_text'] = df['comment_text'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',' ')
```

```python
# Replace numbers with ' '
df['comment_text'] = df['comment_text'].str.replace(r'\d+(\.\d+)?', ' ')
```

```python
# Replace money symbols with ' ' 
df['comment_text'] = df['comment_text'].str.replace(r'Â£|\$', ' ')
```

```python
# Replace 10 digit phone numbers (formats include paranthesis, spaces, no spaces, dashes) with ' '
df['comment_text'] = df['comment_text'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$',' ')
```

```python
import string
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df['comment_text'] = df['comment_text'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in stop_words))
```

```python
# New column (clean_length) after puncuations,stopwords removal
df['clean_length'] = df.comment_text.str.len()
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
#Lemmatization
lem=WordNetLemmatizer()
df['comment_text'] = df['comment_text'].apply(lambda x: ' '.join(lem.lemmatize(t) for t in x.split()))
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, log_loss
```

```python
# Split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from skmultilearn.problem_transform import BinaryRelevance
from skmultilearn.problem_transform import ClassifierChain
```

```python
def build_model(model,estimator,x_train,y_train,x_test,y_test):
    clf=estimator(model)
    clf.fit(x_train,y_train)
    clf_pred=clf.predict(x_test)
    acc=accuracy_score(y_test,clf_pred)
    ll=log_loss(y_test,clf_pred.toarray())
    result= {'accuracy':acc,'log_loss':ll}
    print(classification_report(y_test,clf_pred))
    return result
```

```python
B_r_mnb=build_model(MultinomialNB(),BinaryRelevance,x_train,y_train,x_test,y_test)
```

```python
B_r_gnb=build_model(GaussianNB(),BinaryRelevance,x_train,y_train,x_test,y_test)
```
